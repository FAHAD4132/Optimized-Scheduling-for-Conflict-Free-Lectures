## Graph Coloring Problem
First to solve the problem of design and implement a schedule of the computer science department for the next semester we need represent the lecture scheduling task as a graph coloring problem (GCP).

The graph coloring problem is a problem of assigning colors to the vertices of a graph in such a way that no two adjacent vertices have the same color. This is also called the vertex coloring problem. The graph coloring problem is a special case of graph labeling, which is an assignment of labels to elements of a graph subject to certain constraints.

The graph coloring problem has many applications in various fields, such as design, timetable, sudoku, and map coloring. For example, the graph coloring problem can be used to schedule exams, assign frequencies to radio stations, or color political maps.

The graph coloring problem can be formulated as follows: Given a graph G = (V, E) and a set of m colors, find a way to assign a color to each vertex in V such that no two vertices connected by an edge in E have the same color. Alternatively, find the minimum number of colors needed to color the graph, which is called the chromatic number of the graph.

The graph coloring problem is a hard problem to solve, as it belongs to the class of NP-complete problems. This means that there is no known efficient algorithm that can find the optimal solution for any graph in polynomial time. Therefore, many heuristic algorithms and approximation algorithms have been developed to find good solutions for the graph coloring problem, such as backtracking, greedy, and metaheuristic algorithms.

## Metaheuristics
Metaheuristics are high-level strategies or heuristics that are designed to find, generate, tune, or select a heuristic (partial search algorithm) that may provide a sufficiently good solution to an optimization problem or a machine learning problem, especially with incomplete or imperfect information or limited computation capacity. Metaheuristics sample a subset of solutions which is otherwise too large to be completely enumerated or otherwise explored. Metaheuristics may make relatively few assumptions about the optimization problem being solved and so may be usable for a variety of problems.

Metaheuristics can be classified according to different criteria, such as the type of search strategy, the nature of the solution space, the number of solutions maintained, and the degree of parallelism.

Some examples of metaheuristics algorithms are:
* Hill climbing: This algorithm starts from a random solution and tries to improve it by moving to a neighboring solution that is better than the current one. It repeats this process until it reaches a local optimum, which is a solution that is better than all its neighbors, but not necessarily the best overall solution. Hill climbing can get stuck in local optima and miss the global optimum, which is the best solution in the entire search space.

* Simulated annealing: This algorithm is similar to hill climbing, but it also allows some moves to worse solutions in order to escape from local optima. The probability of accepting a worse solution decreases as the algorithm progresses, mimicking the process of cooling a metal. Simulated annealing can find better solutions than hill climbing, but it requires more time and tuning of parameters.

* Tabu search: This algorithm is an extension of hill climbing that keeps track of a list of forbidden moves, called tabu list, to avoid revisiting the same solutions. The tabu list is updated dynamically based on the recent history of the search. Tabu search can explore more diverse regions of the search space and avoid getting trapped in local optima.

* Genetic algorithms: This algorithm is inspired by the process of natural evolution. It starts with a population of random solutions and applies genetic operators, such as selection, crossover, and mutation, to generate new solutions. The solutions with higher fitness values are more likely to survive and reproduce. Genetic algorithms can find good solutions for complex and nonlinear problems, but they also require more computation and parameter settings.



### Local search algorithms
Local search algorithms are a type of optimization algorithms that are used to find the best solution to a problem by starting from an initial solution and making small changes to it until a desired goal is reached. Local search algorithms are useful for solving problems that are too hard or too expensive to find an exact solution. 

Local search has two limitations:

* It can frequently become trapped at local optima or stuck at plateaus (parts of the search space that have the same value as the objective function)
* It can keep revisiting the same solution or set of solutions (this we call the cycling problem)

We see that both problems can occur even in one-dimensional search spaces:



This is where Tabu search (TS) plays a key role. It was specifically designed to avoid these problems.

### Tabu search
Tabu search is a technique that tries to find the best solution to a problem by exploring different possibilities and avoiding repeating the same ones. 

It works by following these steps:

- Start with a random or given solution and evaluate how good it is using an objective function.
- Generate some similar solutions by making small changes to the current solution. These are called neighbors.
- Choose the best neighbor as the new current solution, even if it is worse than the previous one. This way, the search can escape from local optima, which are solutions that are better than their neighbors but not the best overall.
- Add the current solution to a list of forbidden solutions, called the tabu list. This prevents the search from going back to the same solutions and creating loops.
- Repeat the previous steps until a stopping condition is met, such as a maximum number of iterations, a time limit, or a target value.
- Return the best solution found so far.

Tabu search is useful for solving complex problems that have many possible solutions and where finding the exact optimal solution is too hard or too costly. 

Some examples of problems that can be solved by tabu search are:

- Traveling salesman problem: Find the shortest route that visits a set of cities and returns to the starting point.
- Quadratic assignment problem: Assign a set of facilities to a set of locations, minimizing the cost of transportation and communication between them.
- Knapsack problem: Choose a subset of items that maximizes the total value and fits within a given weight limit.
- Graph coloring problem: Assign a color to each node of a graph, such that no two adjacent nodes have the same color, using the minimum number of colors.

#### Design a tabu search algorithm:

Step 1: select an initial solution $s$ in $S$ (set of solutions) using greedy
* $s_{c} = s; k = 0;$

Step 2: $k = k+1;$ compute $N(s,k)$ such that
* Tabu moves are not considered
* Aspiration criterion may be applicable

Step 3: select the best neighbor solution $s'$ in $N(s,k); s = s';$

Step 4: if $f(s) < f(s_{c})$ then $s_{c} = s;$

Step 5: update the tabu list T and the aspiration criterions.

Step 6: if the stop condition is not reached go to step 2

```python
# Define a function to perform a greedy coloring for a graph
# A greedy coloring assigns the smallest available color to each node
# in the order of the nodes in the graph
def greedy_coloring(graph, timeslots):
    # Create an empty dictionary to store the coloring
    coloring = {}
    # Loop through the nodes of the graph
    for node in graph.nodes():
        # Loop through the colors
        for color in range(timeslots):
            # Assign a value of -1 to the node which means the node is not colored yet
            coloring[node] = -1
            # Find the colors of the neighbors of the current node
            adjacent_colors = {coloring.get(neighbor) for neighbor in graph.neighbors(node)}
            # Assign the smallest color that is not in the adjacent colors
            if color not in adjacent_colors:
                # Assign the color to the node
                coloring[node] = color
                break        
    # Return the coloring
    return coloring
```

```python
# Define a function to find the uncolored nodes in a coloring
# An uncolored node has a color of -1
def find_uncolored(coloring):
    # Create an empty list to store the uncolored vertices
    uncolored = []
    # Loop through the node and their colors
    for node, color in coloring.items():
        # If the color is -1, append the node to the list
        if color == -1:
            uncolored.append(node)
    # Return the list
    return uncolored
```

```python
# Define a function to calculate the degree of a node in a graph
# The degree of a node is the number of its neighbors
def degree(node, graph):
    # Return the length of the neighbor list of the node
    return len(set(graph.neighbors(node)))
```

```python
# Define a function to calculate the total degree of the uncolored nodes in a coloring
# The total degree is the sum of the degrees of the uncolored nodes
def total_degree(coloring, graph):
    # Initialize the total degree to zero
    total = 0
    # Loop through the uncolored vertices
    for node in find_uncolored(coloring):
        # Add the degree of the vertex to the total
        total += degree(node, graph)
    # Return the total degree
    return total
```

```python
# Define a function to find the best neighbor of a coloring
# A neighbor of a coloring is obtained by changing the color of one uncolored node
# The best neighbor is the one that has the lowest total degree and is valid and not in the tabu list
def best_neighbor(coloring, tabu_list, timeslots, graph):
    # Initialize the best neighbor and the best objective value to None
    best_neighbor = None
    best_objective = None
    # Loop through the uncolored nodes
    for node in find_uncolored(coloring):
        # Loop through the available colors
        for color in range(timeslots):
            # Copy the coloring and assign the color to the node
            neighbor = coloring.copy()
            neighbor[node] = color
            # Check if the neighbor is valid and not in the tabu list
            if is_valid(neighbor, graph, timeslots) and neighbor not in tabu_list:
                # Calculate the objective value of the neighbor
                objective = total_degree(neighbor, graph)
                # If the objective value is better than the best objective value, update the best neighbor and the best objective value
                if best_objective is None or objective < best_objective:
                    best_neighbor = neighbor
                    best_objective = objective
    # Return the best neighbor
    return best_neighbor
```

```python
# Define a function to perform a tabu search for a graph coloring problem
# A tabu search is a metaheuristic that explores the neighborhood of a solution
# and avoids revisiting recently visited solutions by using a tabu list
def tabu_search(graph, timeslots, max_iter, tabu_size):
    # Start with a greedy coloring as the current solution
    current = greedy_coloring(graph, timeslots)
    # Initialize the best coloring and the best objective value to the current coloring and its objective value
    best = current
    best_objective = total_degree(current, graph)
    # Initialize the tabu list to an empty list
    tabu_list = []
    # Initialize the iteration counter to zero
    iter = 0
    # Loop until the iteration counter reaches the maximum number of iterations or a feasible coloring is found
    while iter < max_iter and best_objective > 0:
        # Find the best neighbor of the current coloring
        neighbor = best_neighbor(current, tabu_list, timeslots, graph)
        # If no neighbor is found, break the loop
        if neighbor is None:
            break
        # Calculate the objective value of the neighbor
        objective = total_degree(neighbor, graph)
        # If the objective value is better than the best objective value, update the best coloring and the best objective value
        if objective < best_objective:
            best = neighbor
            best_objective = objective
        # Update the current coloring to the neighbor
        current = neighbor
        # Append the current coloring to the tabu list
        tabu_list.append(current)
        # If the tabu list exceeds the tabu size, remove the oldest element
        if len(tabu_list) > tabu_size:
            tabu_list.pop(0)
        # Increment the iteration counter
        iter += 1
    # Return the best coloring
    return best
```

```python
# Define a function to generate graph 
# This function generates a random graph with the given number of nodes and edge probability
def graph_generator(n_nodes, edges_probability=0.1):
    # Create a random graph with the given number of nodes
    G = nx.Graph()
    G.add_nodes_from(range(n_nodes))
    # Add edges to the graph with a probability of edges_probability for each pair of nodes
    for i in range(n_nodes):
        for j in range(i + 1, n_nodes):
            if random.random() < 0.1:
                G.add_edge(i, j) 
    # Return the generated graph
    return G
```

```python
# This function applies the tabu search algorithm to find a near-optimal coloring of a graph
# timeslots is the number of colors to use
# max_iter is the number of iterations to run the algorithm
# tabu_size is the size of the tabu list that stores the forbidden moves
# print_result is to print the result of tabu search algorithm if it equal to True
# graph is the networkx graph object to be colored
def mean(timeslots, graph, max_iter=100, tabu_size=10, print_result=True):
    # Perform the tabu search and get the result
    result = tabu_search(graph, timeslots, max_iter, tabu_size)
    # if print_result equal to True print the result of tabu search algorithm
    if print_result:
        # Print the result and check its validity and quality
        print('Coloring: ', result)
        # Check if the result is a valid coloring, i.e. no two adjacent nodes have the same color
        print('Is feasible ?: ', is_valid_result(result, graph, timeslots))
        # Check how many colors are used in the result
        print('Number of used color K: ', len(set(result.values())))
        # Create a color map for the nodes based on the result
        color_map = [result[node] for node in graph.nodes()]
        # Draw the graph with the color map and the node labels
        nx.draw(graph, node_color=color_map, with_labels=True, font_weight='bold')
        # Show the graph
        plt.show()
    # Return the result
    return result
```

```python
# This function prints the neighbors of each node in a graph
def show_node_neighbors(graph):
    # Create an empty dictionary to store the nodes and their neighbors
    nodes = {}
    # Loop over all the nodes in the graph
    for node in graph.nodes(): 
        # Create an empty list to store the neighbors of the current node
        Node = []
        # Loop over all the neighbors of the current node
        for neighbor in graph.neighbors(node):
            # Append the neighbor to the list
            Node.append(neighbor)
        # Assign the list of neighbors to the node as a value in the dictionary
        nodes[node] = Node
    # Print the dictionary of nodes and their neighbors
    print(nodes)
```

```python
# This function creates a pandas dataframe that shows the schedule of lectures based on the result of the coloring algorithm
def schedule(result, timeslots):
    # Create a list of timeslots (color) as strings with the format "T" followed by the color number
    times = [("T" + str(time)) for time in result.values()]
    # Create a list of unique timeslots (color) as strings with the same format as above
    times_K = [("T" + str(time)) for time in set(result.values())]
    # Create a list to represent the timeslots (color) as strings with the same format as above
    times_columns = [("T" + str(time)) for time in range(timeslots)]
    # Create a list of lectures as strings with the format "L" followed by the node number
    lectures = [("L" + str(lecture)) for lecture in result.keys()]
    # Create an empty list to store the data for the dataframe
    data = []
    # Loop over the lectures
    for l in range(len(lectures)):
        # Create an empty list to store the timeslot for the current lecture
        LT = []
        # Loop over the unique timeslots
        for t in times_columns:
            # If the timeslot matches the one assigned to the lecture, append it to the list
            if t == times[l]:
                LT.append(t)
            # Otherwise, append a blank space to the list
            else:
                LT.append("  ")
        # Append the list of timeslots for the lecture to the data list
        data.append(LT)
    # Create a dataframe from the data list with the lectures as the index and the unique timeslots as the columns
    Data = pd.DataFrame(data, index=lectures, columns=times_columns)
    # Return the dataframe
    return Data
```

```python
# This function creates a pandas dataframe that shows the week schedule of lectures based on the result of the coloring algorithm
def schedule_of_week(result1, result2, result3, result4, result5):
    
    # Create a list of timeslots (color) as strings with the format "T" followed by the color number, repet for all result of the coloring algorithm
    times1 = [("T" + str(time)) for time in result1.values()]
    times2 = [("T" + str(time)) for time in result2.values()]
    times3 = [("T" + str(time)) for time in result3.values()]
    times4 = [("T" + str(time)) for time in result4.values()]
    times5 = [("T" + str(time)) for time in result5.values()]
    
    # Create a list to represent the days of a week
    columns = ["Saturday", "Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday"]
    
    # Create a list of lectures as strings with the format "L" followed by the node number
    lectures = [("L" + str(lecture)) for lecture in result1.keys()]
    
    # Create an empty list to store the data for the dataframe
    data = np.zeros((len(lectures), 7)).astype(str)
    
    # Loop over the lectures
    for i in range(len(lectures)):
        data[i][0] = " "
        data[i][1] = times1[i]
        data[i][2] = times2[i]
        data[i][3] = times3[i]
        data[i][4] = times4[i]
        data[i][5] = times5[i]
        data[i][6] = " "
        
    # Create a dataframe from the data list with the lectures as the index and the days of the week as the columns
    Data = pd.DataFrame(data, index=lectures, columns=columns)
    
    # Return the dataframe
    return Data
```

There are cases where it is impossible to find a feasible solution in tabu search. This can happen when the problem is infeasible, meaning that there is no solution that satisfies all the constraints or requirements of the problem. For example, if the problem is to color a graph with k colors such that no two adjacent vertices have the same color, and the graph has a clique of size k+1, then there is no feasible solution. A clique is a subset of vertices that are all connected to each other by edges.

Another reason why tabu search may fail to find a feasible solution is when the initial solution is infeasible and the neighborhood function does not generate any feasible neighbors. A neighborhood function is a way of generating similar solutions by making small changes to the current solution. If the neighborhood function is too restrictive or too random, it may not be able to produce any feasible solutions. For example, if the problem is to assign $n$ tasks to $m$ workers, and the initial solution assigns more than $\dfrac{n}{m}$ tasks to some workers, then the neighborhood function may not be able to reduce the workload of those workers without violating other constraints. 

To avoid these situations, it is important to check the feasibility of the problem and the initial solution before applying tabu search. It is also advisable to use a neighborhood function that can explore different regions of the solution space and generate diverse and feasible solutions.

Some of way to overcome this issue:

**Dynamic Time Slot Adjustments:**
* Allow for dynamic adjustments to the time slots, where the algorithm can iteratively modify the schedule by swapping or adjusting time slots until a feasible solution is found. This helps in exploring the solution space more effectively.

5. **Memory Mechanism (Tabu List):** Leverage the tabu list to guide the search, preventing the algorithm from revisiting previously explored solutions.

6. **Iteration and Evolution (GA):** Continue the process through multiple iterations of genetic evolution and tabu search refinement until a satisfactory solution is found or a convergence criterion is met.

This hybrid approach aims to combine the strengths of both algorithms to efficiently explore the solution space, maintain diversity, and refine solutions to meet the specific constraints of the scheduling problem. Experimentation and parameter tuning will be crucial to optimizing the performance of this hybrid metaheuristic.
